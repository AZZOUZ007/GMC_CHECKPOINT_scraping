```python
from bs4 import BeautifulSoup
import requests
```

```python
#1.1) Write a function to Get and parse html content from a Wikipedia page

url = input("Collez le lien ici ==> ")
def content_parser(url):
    code_source = requests.get(url).content
    soup = BeautifulSoup(code_source, 'html.parser')
    
    return soup
content_parser(url)
```

```python
thin = content_parser(url).select('a')

i = 0
for element in thin:
    link = element.get('href')
    if link is not None:
        if link[:5] == 'https':
            if link[0:16] in ['https://en.wiki','https://fr.wiki']:
                i += 1
                print(f'{i} link')
        if link[:5] != 'https' and link[0] != "#":
            i += 1
            print(f"{i} - https://www.wikipedia.org{link}")
```
